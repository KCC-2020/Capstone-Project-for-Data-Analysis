# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here

import matplotlib.pyplot as plt
import seaborn as sns

# fecthing data from the sql table
QUERY= '''SELECT * FROM master   '''
df = pd.read_sql_query(QUERY,conn)

# Select data from the ConvertedComp column and plot the histogram
sns.histplot(df['ConvertedComp'])
plt.title('Distribution of converted compensation')
plt.ylabel('Number of Survey responders')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here

sns.boxplot(x=df['Age'])
plt.title('Boxplot of Age distribition in the Survey')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sns.scatterplot(x=df['Age'], y=df['WorkWeekHrs'], data=df)
plt.title('Distribution of Age in relation with WorkWeekHrs')


```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
sns.scatterplot(x=df['WorkWeekHrs'], y=df['CodeRevHrs'], size=df['Age'], data=df)
plt.title('Distribution of WorkWeekHrs and CodeRevHrs in relation with Age')
plt.legend()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here

#fetch data
QUERY = """ SELECT * FROM DatabaseDesireNextYear """
rows = conn.execute(QUERY).fetchall()
df_pie = pd.read_sql_query(QUERY,conn)

# create the dataframe of top5 database
df_top5 = df_pie['DatabaseDesireNextYear'].value_counts().to_frame().reset_index().head()
df_top5.reset_index(drop=True)

#plot data
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'pink']
plt.figure(figsize=(12,6))

df_top5['DatabaseDesireNextYear'].plot(kind='pie', 
                                        startangle = 90,
                                        labels= None,
                                        shadow=True,
                                        autopct ='%1.1f%%', 
                                        pctdistance=1.12,
                                        colors=colors_list)

plt.title('Top5 Databases desired by respondents', y=1.12) 
plt.axis('equal')
plt.legend(labels=df_top5['index'], loc='upper left')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# select the dataframe
QUERY='''SELECT WorkWeekHrs,CodeRevHrs,Age from master WHERE Age between 30 and 35'''
group_age_df = pd.read_sql_query(QUERY,conn)

# plot data 
plt.bar(x=group_age_df['Age'], height=group_age_df['WorkWeekHrs'], color='lightgreen')
plt.bar(x=group_age_df['Age'], height=group_age_df['CodeRevHrs'],color='red')
plt.title('Stacked chart of median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Number of Hours worked')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

# select the dataframe
QUERY='''SELECT ConvertedComp,Age from master WHERE Age between 45 and 60'''
age60_df = pd.read_sql_query(QUERY,conn)

#group the data 
data_age60 = age60_df.groupby(['Age'])['ConvertedComp'].sum()
data_age60 = data_age60.to_frame()

#plot the data
data_age60.plot(kind='line', color='red')
plt.title("The median ConvertedComp for all ages from 45 to 60")
plt.ylabel("Compensation Paid (X 1000000 US dollars)")
plt.xticks(rotation=45)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

#fetch data
QUERY='''SELECT MainBranch FROM master '''
branch_df = pd.read_sql_query(QUERY,conn)
branch_df = branch_df.value_counts().to_frame()

# plot the data
labels=list(df['MainBranch'])
branch_df.plot(kind='barh')
plt.title('Distribution of respondents by MainBranch')
plt.legend(labels)
plt.xlabel('Number of Responders')

```
